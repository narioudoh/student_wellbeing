```sql

SELECT 
    Dietary_Habits,
    Sleep_Duration,
    Financial_Stress,
    SUM(COALESCE(Depression, 0)) AS depressed_count,
    COUNT(1) AS total_count,
    CAST(SUM(COALESCE(Depression, 0)) * 100.0 / COUNT(1) AS DOUBLE) AS depression_rate
FROM analysis.dev.student
GROUP BY Dietary_Habits, Sleep_Duration, Financial_Stress
ORDER BY depression_rate DESC;
SELECT 
    Dietary_Habits,
    Sleep_Duration,
    Financial_Stress,
    SUM(Depression) AS depressed_count,
    COUNT(*) AS total_count,
    (SUM(Depression) * 100.0 / COUNT(*)) AS depression_rate
FROM analysis.dev.student
GROUP BY Dietary_Habits, Sleep_Duration, Financial_Stress
ORDER BY depression_rate DESC;
```
