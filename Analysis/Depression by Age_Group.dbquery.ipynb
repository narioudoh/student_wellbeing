```sql
-- Depression rate by age groupn --
SELECT 
    CASE 
        WHEN Age < 20 THEN '<20'
        WHEN Age BETWEEN 20 AND 25 THEN '20-25'
        WHEN Age BETWEEN 26 AND 30 THEN '26-30'
        ELSE '30+' 
   END AS age_group,
    SUM(Depression) AS depressed_students,
    COUNT(*) AS total,
    (SUM(Depression) * 100 / COUNT(*)) AS depression_rate
FROM student
GROUP BY age_group;

```
