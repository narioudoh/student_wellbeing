```sql
SELECT 
    Suicidal_Thoughts AS had_suicidal_thoughts,
    SUM(
        CASE 
            WHEN Suicidal_Thoughts = 'Yes' THEN 1 
            ELSE 0 
        END
    ) AS Attempted_Suicide,
    COUNT(*) AS total,
    (
        SUM(
            CASE 
                WHEN Suicidal_Thoughts = 'Yes' THEN 1 
                ELSE 0 
            END
        ) * 100.0
        / count(suicidal_thoughts)
    ) AS suicidal_rate
FROM student
GROUP BY Suicidal_Thoughts
```
