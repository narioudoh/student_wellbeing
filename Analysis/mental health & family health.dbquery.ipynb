```sql
SELECT 
    Family_History_of_Mental_Illness,
    SUM(Depression) AS depressed_count,
    SUM(CASE WHEN suicidal_thoughts = 'Yes' THEN 1 ELSE 0 END) AS suicidal_thoughts_count,
    COUNT(*) AS total_count,
    (SUM(Depression) * 100.0 / COUNT(*)) AS depression_rate,
    (SUM(CASE WHEN suicidal_thoughts = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*)) AS suicidal_thoughts_rate
FROM analysis.dev.student
GROUP BY Family_History_of_Mental_Illness
ORDER BY depression_rate DESC
```
